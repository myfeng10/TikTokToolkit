```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import pyperclip
import json
```

```python
url = 'https://script.tokaudit.io/'

links_to_search = [
    'https://www.tiktok.com/@unclimatechange/video/7332929544967458081?lang=en',
    'https://www.tiktok.com/@unclimatechange/video/7303599892922977569?lang=en',
    'https://www.tiktok.com/@unclimatechange/video/7309605013410647329?lang=en',
    # 'https://www.tiktok.com/@unclimatechange/video/7197478740513934598?lang=en'
    # 'https://www.tiktok.com/@unclimatechange/video/7350642104391044385?lang=en'
]
```

```python
driver = webdriver.Firefox()
driver.get(url)

text={}
for link in links_to_search:
    search_bar = driver.find_element(By.CSS_SELECTOR, 'input[placeholder="Insert a Tiktok video..."]')
    search_bar.clear()
    search_bar.send_keys(link)
    search_bar.send_keys(Keys.RETURN)
    
    start = driver.find_element(By.XPATH, "//button[@type='button'][contains(@class,'bg-blue-600')]")
    start.click()

    elements = driver.find_elements(By.XPATH, "//div[contains(@class,'bg-red-500')][contains(text(),'could not download subtitles ')]")
    if elements:
        print("Could not download subtitles for:", link)
        continue

    copy = WebDriverWait(driver, 20).until(
    EC.visibility_of_element_located((
        By.XPATH, "//button[contains(@class,'bg-pink-600')][contains(text(),'Copy')]"
    ))
    )
    copy.click()

    copied_text = pyperclip.paste()
    text[link] = copied_text
    print(link)
    print(copied_text)

driver.quit()
```

```python
print(json.dumps(text, indent=4))
```
